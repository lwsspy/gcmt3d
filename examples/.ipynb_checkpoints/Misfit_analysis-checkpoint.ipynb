# Misfit Analysis of a simple amplitude and phase shifted signal

We found that the amplitude/energy is not well recovered by waveform inversion 
if one cannot affect the phase.


```python
from scipy.signal.windows import hann as taper
from scipy.signal import hilbert
# from scipy.optimize import 
import numpy as np

# 
def envelope(array):
    return np.abs(hilbert(array))


times = np.linspace(0, 2*np.pi,  1000)
data  = np.sin(3*times) * taper(len(times))
model = np.array([.5])

def forward(a):
    return a * np.sin(3*times + np.pi*0.25) * taper(len(times))
    
def compute_cost_L2(a):
    return 0.5 * np.sum((forward(a) - data)**2)/np.sum(data**2)

def derivative_L2(a):
    return (forward(a) - data) * (forward/a)

def compute_cost_ln(a):
    return 0.5 * (0.5 * np.log( np.sum(forward(a)**2) / np.sum(data**2) ))**2

def compute_cost_env(a):
    return 0.5 * np.sum((envelope(forward(a)) - envelope(data))**2)/np.sum(envelope(data)**2)

def compute_cost(a, lmb=0.1):
    return (1-lmb) * compute_cost_L2(a) + lmb * compute_cost_env(a)

def derivative_ln(a):
    return (1/a)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(4,4), facecolor='w')
plt.subplot(2,1,1)
plt.plot(times,data)
plt.plot(times,forward(model))

plt.subplot(2,1,2)
plt.plot(times,envelope(data))
plt.plot(times,envelope(forward(model)))
plt.savefig("problem.png", dpi=300)
```

To get the ratio in energy between the two signal 

```python
ratio_env = np.sqrt(np.sum(envelope(data)**2)/np.sum(envelope(forward(model))**2))
ratio_signal = np.sqrt(np.sum(data**2)/np.sum(forward(model)**2))
print(ratio_env)
print(ratio_signal)
```

## Checking Amplitude Misfit

```python
from functools import partial
models = np.linspace(0.1, 2.0, 100)

def plot_misfit(m, costfunc, *args, **kwargs):
    
    # Plot cost curve
    cost = np.array([costfunc(_m) for _m in m])
    cmax = np.max(cost)
    line, = plt.plot(m, cost/cmax, *args, **kwargs)
    
    # Plot minimum cost and expected cost
    pos = np.argmin(cost)
   
    plt.plot(m[pos], cost[pos]/cmax, 'o', color=line.get_color(), markerfacecolor='none')
    plt.plot(1, costfunc(1)/cmax,'x', color=line.get_color(),  markerfacecolor='none')


    

plt.figure(facecolor='w')
plot_misfit(models, compute_cost_L2, label="$C_{L^2}$")
plot_misfit(models, compute_cost_ln, label="$C_{\ln}$")
plot_misfit(models, compute_cost_env, label="$C_{e}$")
plot_misfit(models, partial(compute_cost, lmb=0.5), label="$C$")
plt.legend(loc='lower right', ncol=1)
plt.savefig("costfunction_amp.png", dpi=300)
```

### Checking different values for $\lambda$

```python
lmbds = np.linspace(0.1, 1.0, 10)

plt.figure(facecolor='w')
for lmb in lmbds:
    plot_misfit(models, partial(compute_cost, lmb=lmb), label=f"$\lambda = {lmb:3.1f}$")
plt.legend(loc='lower right', ncol=1)
plt.xlim(0, 2.25)
plt.savefig("lambda1D.png", dpi=300)
```

## Checking both time and amplitude

First we have to setup all the functions that we need to cmpute a grid search.

```python
from scipy.signal.windows import hann as taper
# from scipy.optimize import 
import numpy as np

from scipy.signal import hilbert

def envelope(array):
    return np.abs(hilbert(array))


times = np.linspace(0, 2*np.pi,  1000)
data  = np.sin(3*times) * taper(len(times))
model = np.array([0.5])


def forward(a):
    a, shift = a
    return a * np.sin(3*times + np.pi*shift) * taper(len(times))    
    
def compute_cost_L2(a):
    return 0.5 * np.sum((forward(a) - data)**2)/np.sum(data**2)

def compute_cost_ln(a):
    return (0.5 * np.log( np.sum(forward(a)**2) / np.sum(data**2) ))

def compute_cost_env(a):
    return 0.5 * np.sum((envelope(forward(a)) - envelope(data))**2)/np.sum(envelope(data)**2)

def compute_cost_env_ratio(a):
    return np.sqrt(np.sum(envelope(data)**2)/np.sum(envelope(forward(a))**2))

def compute_cost(a, lmb=0.5):
    return (1-lmb) * compute_cost_L2(a) + lmb * compute_cost_env(a)
```

Then, we compute the cost function for each scenario:

```python
pis = np.linspace(np.pi/32, np.pi/8, 100)
a_d  = np.linspace(0.02, 2, 100)

# 
xa, xp = np.meshgrid(a_d, pis)

# 
cost_L2 = np.zeros_like(xa)
cost_ln = np.zeros_like(xa)
cost_env = np.zeros_like(xa)
cost_env_ratio = np.zeros_like(xa)
cost_C = np.zeros_like(xa)

models = []
indeces = []
for _i, _a in enumerate(a_d):
    for _j, _p in enumerate(pis):
        indeces.append([_j, _i])
        _m = np.array([_a, _p])
        cost_L2[ _j, _i] = compute_cost_L2(_m)
        cost_ln[ _j, _i] = compute_cost_ln(_m)
        cost_env[ _j, _i] = compute_cost_env(_m)
        cost_env_ratio[ _j, _i] = compute_cost_env_ratio(_m)
        cost_C[ _j, _i] = compute_cost(_m)
```

```python
a_d
```

```python
def min2D(arr2D):
    return np.where(arr2D == np.min(arr2D))
```

```python
fig = plt.figure(figsize=(15,4), facecolor='w')


def plot_misfit(a,p, cost, ylabel=False):
    ix, fx = 0, len(a)
    iy, fy = 0, len(p)
    plt.pcolormesh(
        xa[ix:fx, iy:fy], xp[ix:fx, iy:fy], cost[ix:fx, iy:fy], shading='auto')
    plt.colorbar(orientation='horizontal', pad=0.25)
    pp, pa = min2D(cost)
    print(pp,pa)
    plt.plot(a_d[pa], xp[pp], 'wo', markerfacecolor='none', clip_on=False)
    m = [1 , 0]
    plt.plot(*m, 'wx', markerfacecolor='none')
    plt.xlabel("Amplitude")
    plt.xlim((0.9, 1.1))
    plt.ylim((-0.2, 0.2))
    if ylabel:
        plt.ylabel("Phase-shift")
    
    model = a_d[pa][0], xp[pp][0][0]
    return model


solutions = []

plt.subplot(1,6,1, facecolor='grey')
solutions.append(plot_misfit(xa, xp, cost_L2, ylabel=True))
plt.title(r"$C_{L^2} = \frac{1}{2}\sum\|d - s\|^2 \left/ \sum d^2 \right.$")


plt.subplot(1,6,2, facecolor='grey')
solutions.append(plot_misfit(xa, xp, cost_ln))
plt.title(r"$C_{\ln} = \frac{1}{2}\ln \sum d^2 \left/ \sum s^2 \right.$")


plt.subplot(1,6,3, facecolor='grey')
solutions.append(plot_misfit(xa, xp, cost_env))
plt.title(r"$C_{e} = \frac{1}{2} \sum \| e(d) -  e(s) \|^2 \left/ \sum e^2(d)^2 \right.$")


plt.subplot(1,6,4, facecolor='grey')
solutions.append(plot_misfit(xa, xp, cost_env_ratio))
plt.title(r"$C_{e} = \frac{1}{2} \ln e^2(d) \left/ e^2(s)  \right.$")

plt.subplot(1,6,5, facecolor='grey')
solutions.append(plot_misfit(xa, xp,  0.5 * (cost_L2/np.max(cost_L2) + cost_ln/np.max(cost_ln))))
plt.title(r"$\frac{1}{2} (C_{L^2} + C_{\ln})$")

plt.subplot(1,6,6, facecolor='grey')
solutions.append(plot_misfit(xa, xp, cost_C/np.max(cost_C)))
plt.title(r"$\frac{1}{2} (C_{L^2} + C_{e})$")

# Adjustments and colorbar
# fig.subplots_adjust(right=0.8, wspace=0.3)
# cbar_ax = fig.add_axes([0.825, 0.1, 0.01, 0.8])
# fig.colorbar(p, cax=cbar_ax)

plt.savefig("minimum_finding_best_model.png", dpi=300)
```

```python
labels = ["$C_{L^2}$", "$C_{\ln}$","$C_{e}$","$C_{L^2} + C_{\ln}$","$C_{L^2} + C_{e}$"]

plt.figure(facecolor='w')
plt.plot(times, data, 'k', label="data")
for m, lab in zip(solutions, labels):
    print(m)
    plt.plot(times, forward(m), label=lab)
    
plt.legend(loc='lower right', ncol=1)

plt.savefig('solutions.png', dpi=300)
```

### Optimizing for lambda

First, we compute the the cost functions 

```python
pis = np.linspace(np.pi/16, np.pi/8, 100)
a_d  = np.linspace(0.01, 2, 200)
lmbs = np.linspace(0.0, 2.0, 3)

# Create mesh for plotting
xa, xp = np.meshgrid(a_d, pis)

# 
cost_L2 = np.zeros_like(xa)
cost_env = np.zeros_like(xa)
cost_ln = np.zeros_like(xa)

models = []
indeces = []


for _i, _a in enumerate(a_d):
    for _j, _p in enumerate(pis):
        _m = np.array([_a, _p])
        cost_L2[ _j, _i] = compute_cost_L2(_m)
        cost_env[ _j, _i] = compute_cost_env(_m)
        cost_ln[ _j, _i] = compute_cost_ln(_m)
```

```python
l2_max = np.max(cost_L2)
ln_max = np.max(np.abs(cost_ln))
env_max = np.max(cost_env)
```

The, we get the minimum for each lambda. 

```python
cost_lmd = np.zeros((len(lmbs), *np.shape(xa)))

for _k, _lmb in enumerate(lmbs):
    cost_lmd[_k ,:,:] = cost_L2/l2_max + _lmb * cost_env/env_max

lmbs = np.linspace(0.1, 0.3, 11)

solutions = []
mins = []
for _k, _lmb in enumerate(lmbs):
    cost = cost_L2/l2_max + _lmb * cost_ln/ln_max
    mins.append(np.min(cost))
    pos = min2D(cost)
    solutions.append([a_d[pos[1]], pis[pos[0]]])
```

Then, we plot the solutions

```python
plt.figure(figsize=(9,3),facecolor='w')
ax1 = subplot(2,2,1)
ax11 = subplot(2,2,3)
ax2 = subplot(2,2,2)
ax3 = subplot(2,2,4)
ax1.plot(times, data, 'k', label="data")
ax11.plot(times, envelope(data), 'k', label="data")
ax2.plot(times, data, 'k', label="data")
ax3.plot(times, data, 'k', label="data")

N=2
for _m, _lmb, _min in zip(solutions[::N], lmbs[::N], mins[::N]):

    ax1.plot(times, forward(_m), label=f"$\lambda = {_lmb:4.2f}$")
    ax11.plot(times, envelope(forward(_m)), label=f"$\lambda = {_lmb:3.1f}$")
    ax2.plot(times, forward(_m), label=f"$\lambda = {_lmb:4.2f},\, min={_min:e}$")
    ax3.plot(times, forward(_m), label=f"$\lambda = {_lmb:4.2f}$")
    
ax2.legend(bbox_to_anchor=(1.0,0.5), loc='upper left', ncol=1)
ax1.tick_params(labelbottom=False)
ax2.tick_params(labelbottom=False)
ax1.set_xlim(0, 2*np.pi)
ax11.set_xlim(0, 2*np.pi)
ax2.set_xlim(2.3,2.8)
ax2.set_ylim(0.8,0.95)
ax3.set_xlim(3.25,3.75)
ax3.set_ylim(-0.985,-0.9)

plt.savefig('optimize_lambda.png', dpi=300) 
```
